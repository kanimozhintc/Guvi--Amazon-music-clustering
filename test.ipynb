```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#convert the datatype:
df['release_date']=pd.to_datetime(df['release_date'], errors='coerce')
```

```python
#Feature selection:
features=['danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness',
            'valence', 'tempo', 'duration_ms']
```

```python
X=df[features]
X
```

```python
#Distribution of song Popularity:
plt.figure(figsize=(8, 6))
sns.histplot(df['popularity_songs'], kde=True)
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Song Popularity')
plt.show()
#Insights:
#Many songs in the dataset have popularity = 0, meaning they are not streamed/listened to much at all.
#Many songs have low-to-medium popularity, with noticeable clustering between 20–40.
#Very few songs reach very high popularity (>70).
#The curve drops quickly after ~50, meaning only a small fraction of songs are "hits".
```

```python
#Correlelation:
#Energy & Loudness = 0.73
# Very strongly correlated.
plt.figure(figsize=(8,6))
sns.heatmap(X.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()
```

```python
#Here it shows the song durations has been distributed:
plt.figure(figsize=(8, 6))
sns.histplot(df['duration_ms'] / (1000 * 60), kde=True)
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Song Durations')
plt.show()
#Insights:
#Most songs are between 2 and 5 minutes long.There is a strong peak in this range.
#Some very short tracks (<1 minute).Could be intros,podcasts split into short segments.
#Long songs exist but are rare (>10 minutes).
#The histogram quickly drops off after ~6 minutes.
#Songs lasting 20+ minutes are extremely uncommon (could be live recordings, DJ mixes, or classical pieces).
```

```python
#Scaling the features:
scalar=StandardScaler()
X_scaled=scalar.fit_transform(X)
```

```python
#Apply PCA :
pca=PCA(n_components=5)
X_pca=pca.fit_transform(X_scaled)

print("Shape before PCA:", X_scaled.shape)
print("Shape after PCA:", X_pca.shape)
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total variance explained:", pca.explained_variance_ratio_.sum())
```

```python
#Elbow method---To find the K optimal value:
inertia = []
K = range(1, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    print('K value = ', k, 'Inertia value = ', kmeans.inertia_)
plt.figure(figsize=(8,5))
plt.plot(K, inertia, marker='o', linestyle='--', color='b')
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method - Spotify Songs")
plt.xticks(K)
plt.show()
```

```python
#Apply Kmeans Clustering:
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster']=kmeans.fit_predict(X_scaled)
```

```python
#Evaluate Clusters ---
silhouette_avg = silhouette_score(X_scaled, df['cluster'])
print(f"Silhouette Score: {silhouette_avg:.3f}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=df["cluster"], cmap="viridis")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Spotify Clusters after PCA")
plt.show()

#Insights:
#cluster 0:energy low+high acousticness---Chill Acousticness
#cluster 1:dance+High energy+High tempo+High valence---Party track
#cluster 2:High dance+medium energy+high speechiness+high liveness----rap song,live recordings or spoken word content
```

```python
#To find the average of cluster:
cluster_summary = df.groupby("cluster")[features].mean()
print(cluster_summary)

```

```python
#Distribution Plots (e.g., Danceability, Tempo, Energy):
features_to_plot = ["danceability", "tempo", "energy"]

for feature in features_to_plot:
    plt.figure(figsize=(7,4))
    sns.kdeplot(data=df, x=feature, hue="cluster")
    plt.show()

#Insights for danceability:
#Cluster 0----less danceable songs.
#Cluster 1---- moderately danceable songs.
#Cluster 2---- very danceable songs.
```

```python
cluster_means=df.groupby('cluster')[['danceability', 'energy','acousticness','valence','speechiness','liveness' ]].mean().reset_index()
 #cluster_avg is grouped by cluster
cluster_means.plot(kind='bar', figsize=(10,6))
plt.title("Average Feature Values per Cluster")
plt.xlabel("Cluster")
plt.ylabel("Average Value")
plt.legend(title='Feature')
plt.show()
```
